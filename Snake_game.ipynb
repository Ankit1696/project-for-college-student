```python
# import necessary libraries
import pygame
import random
import sys

# define constants and colors
WIDTH, HEIGHT = 600, 600
BLOCK_SIZE = 20
FPS = 10
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLACK = (0, 0, 0)
UP = 'up'
DOWN = 'down'
LEFT = 'left'
RIGHT = 'right'

# initialize pygame
pygame.init()
# create window object
window = pygame.display.set_mode((WIDTH, HEIGHT))
# create clock object
clock = pygame.time.Clock()

# initialize snake position and direction
snake = [(300, 300), (280, 300), (260, 300)]
direction = RIGHT

# initialize food position
food_x = random.randint(0, (WIDTH - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE
food_y = random.randint(0, (HEIGHT - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE

# initialize score
score = 0

# create font object
score_font = pygame.font.SysFont('consolas', 20)

# main game loop
while True:
    # check for user input
    for event in pygame.event.get():
        # if user presses ESC, exit the game
        if event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
            sys.exit()
        # if user presses arrow keys, change direction
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_UP:
            if direction != DOWN:
                direction = UP
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_DOWN:
            if direction != UP:
                direction = DOWN
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_LEFT:
            if direction != RIGHT:
                direction = LEFT
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_RIGHT:
            if direction != LEFT:
                direction = RIGHT

    # update snake position
    # get the head position
    head_x, head_y = snake[0]
    # calculate the new head position according to direction
    if direction == UP:
        new_head = (head_x, head_y - BLOCK_SIZE)
    elif direction == DOWN:
        new_head = (head_x, head_y + BLOCK_SIZE)
    elif direction == LEFT:
        new_head = (head_x - BLOCK_SIZE, head_y)
    elif direction == RIGHT:
        new_head = (head_x + BLOCK_SIZE, head_y)
    # insert the new head at the front of the list
    snake.insert(0, new_head)
    if score > 15 and score<= 30:
        FPS = 15
    elif score > 30:
        FPS = 25
    # remove the last segment of the list
    snake.pop()

    # check for collision with food
    if snake[0] == (food_x, food_y):
        # increase score
        score += 1
        # add a new segment to the snake
        snake.append((food_x, food_y))
        # generate a new food position
        food_x = random.randint(0, (WIDTH - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE
        food_y = random.randint(0, (HEIGHT - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE
#     else:
#         snake.pop()

    # check for collision with boundaries or itself
    elif (snake[0][0] < 0 or snake[0][0] >= WIDTH or
        snake[0][1] < 0 or snake[0][1] >= HEIGHT or
        snake[0] in snake[1:]):
        # end the game and display game over message
        game_over = score_font.render('Game Over', True, WHITE)
        window.blit(game_over, (WIDTH // 2 - game_over.get_width() // 2, HEIGHT // 2 - game_over.get_height() // 2))
        pygame.display.flip()
        pygame.time.wait(3000)
        sys.exit()

    # clear the window
    window.fill(BLACK)

    # draw the snake
    for x, y in snake:
        pygame.draw.rect(window, RED, (x, y, BLOCK_SIZE, BLOCK_SIZE))

    # draw the food
    pygame.draw.rect(window, WHITE, (food_x, food_y, BLOCK_SIZE, BLOCK_SIZE))

    # display the score
    pygame.display.set_caption(f'Snake Game | Score: {score}')

    # update the display
    pygame.display.flip()

    # control the game speed
    clock.tick(FPS)
```
